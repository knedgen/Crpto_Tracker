```python
import pandas as pd
import requests
import bs4

#gets Coin Market Cap homepage and converts to beautiful soup object
result = requests.get("https://coinmarketcap.com/")
soup = bs4.BeautifulSoup(result.text,'lxml')

#gathers names of top 10 cryptos and puts into a list
names = []
for name in soup.select('.iJjGCS')[8:]:
    names.append(name.text)

#grabs tickers
symbols = []
for syms in soup.select('.gGIpIK')[0:]:
    symbols.append(syms.text)


#gets ranks
ranks = []
for rank in soup.select('.etpvrL')[0::2]:
    ranks.append(rank.text)


#for getting prices    
prices = []
for price in soup.select('.price___3rj7O')[0:]:
    prices.append(price.text)
prices    


# collects the market cap values
caps = []
for cap in soup.select('.kDEzev')[1:20:2]:
    caps.append(cap.text)

data = {'Name':names,
       'Ticker': symbols,
       'Ranks': ranks,
       'Prices': prices,
       'Market Cap': caps}

df = pd.DataFrame(data, columns=['Name','Ticker','Ranks','Prices','Market Cap'])
    
df
```
