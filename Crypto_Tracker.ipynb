```python
import pandas as pd
import requests
import bs4

#gets Coin Market Cap homepage and converts to beautiful soup object
result = requests.get("https://coinmarketcap.com/")
soup = bs4.BeautifulSoup(result.text,'lxml')

#gathers names of top 10 cryptos and puts into a list
names = []
for name in soup.select('.sc-1eb5slv-0.iworPT')[9:]:
    names.append(name.text)

#grabs tickers
symbols = []
for syms in soup.select('.gGIpIK')[9:]:
    symbols.append(syms.text)


#gets ranks
ranks = []
for rank in soup.select('.etpvrL')[0::2]:
    ranks.append(rank.text)


#for getting prices    
prices = []
for price in soup.select('div.sc-131di3y-0.cLgOOr')[3:]:
    prices.append(price.text)    


# collects the market cap values
caps = []
for cap in soup.select('span.sc-1ow4cwt-1.ieFnWP'):
    caps.append(cap.text)

data = {'Name':names,
       'Ticker': symbols,
       'Rank': ranks,
       'Price': prices,
       'Market Cap': caps}

df = pd.DataFrame(data, columns=['Rank','Name','Ticker','Price','Market Cap'])
    
df
```
